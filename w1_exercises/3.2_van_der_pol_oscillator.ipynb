```python
# The equation
# (d^2x)/(dt^2) - u(1-x^2) * dx/dt + x = 0
# is called the Van der Pol oscillator.

# Rewrite the equation to a system of two coupled first-order ODEs.


#a 
# dx/dt = y
# dy/dt = u(1 - x^2)y - x

```

```python
# b
# Solve the equation using the Euler method for u = 0.1, u = 1.0,
# u = 10, u = 100, and u = 250 with initial conditions x(0) = 2 and
# x'(0) = 0. Solve for t ∈ [0, 10u] and use Δt = 0.01.

import numpy as np
import matplotlib.pyplot as plt

def f_derived(x, y, u):
    return u * (1 - x**2) * y - x

def euler_method(u, dt, t_max):
    t_values = np.arange(0, t_max, dt)
    x_values = [2]
    y_values = [0]

    for i in range(1, len(t_values)):
        t_i = t_values[i]
        x_i = x_values[i - 1]
        y_i = y_values[i - 1]
        
        x_new = x_i + dt * y_i
        y_new = y_i + dt * f_derived(x_i, y_i, u)
        
        x_values.append(x_new)
        y_values.append(y_new)

    return t_values, x_values, y_values

dt = 0.01
t_max = 10

u_values = [0.1, 1.0, 10, 100, 250]
initial_conditions = (2, 0)

for u in u_values:
    t, x, y = euler_method(u, dt, t_max*u)
    plt.plot(t, x, label=f'u = {u}')

plt.xlabel('Time')
plt.ylabel('x')
plt.title('Van der Pol Oscillator')
plt.legend()
plt.grid(True)
plt.show()
```
