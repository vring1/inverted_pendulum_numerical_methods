```python
import numpy as np
import scipy.sparse as sp
import scipy.sparse.linalg as splinalg
import matplotlib.pyplot as plt

# Parameters
N = 1000       # Number of grid points
L = 1.0        # Length of the domain
dt = 0.001     # Time step size
dx = L / N     # Spatial step size

# Discretization
x = np.linspace(0, L, N+1)   # Spatial grid points
f = np.exp(-5 * x)           # Initial condition

# Time stepping
steps = 1000
A = sp.diags([1, -2, 1], [-1, 0, 1], shape=(N-1, N-1)).toarray()  # Laplacian matrix
print(A)
for step in range(steps):
    f[1:N] += dt * np.dot(A, f[1:N])
    
# Plot the result
plt.plot(x, f)
plt.xlabel('x')
plt.ylabel('f(x, t)')
plt.title('Time evolution of the 1D heat equation')
plt.show()
```
