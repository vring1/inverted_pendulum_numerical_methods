```python
import scipy
z = np.linspace(0,150,151)
lb = my_lambertw(z)
lb2 = scipy.special.lambertw(z)
import matplotlib.pyplot as plt

plt.plot(z,lb, label="My Lambert")
plt.plot(z,lb2, label="My Lambert")

plt.legend()
plt.show()
```
